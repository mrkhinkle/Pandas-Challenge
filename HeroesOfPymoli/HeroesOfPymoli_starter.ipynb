```python
# Dependencies and Setup
import pandas as pd
import numpy as np


# File to Load (Remember to Change These)
read_file = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(read_file, encoding="utf8")
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#Clean data of rows with missing data
clean_purchase_data = purchase_data.dropna(how="any")
clean_purchase_data.count()


#Do a count of unique values for SN
player_count = len(purchase_data["SN"].unique())

#Build a table displaying the total of unique screen names
player_count_table = pd.DataFrame({"Count of Players":[player_count]})
player_count_table
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Grouping purchase data first by Gender
player_group = purchase_data.groupby(["SN"])

#Pulling an array of unique values for vender by SN and counting them by gender
gender_unique = player_group["Gender"].unique()
gender_count = gender_unique.value_counts()

#create a metric value for user gender totals as a percentage of the total players
gender_percentage = ((gender_count/player_count)*100)

#Create a DF combining gender counts with gender percentage
gender_summary = pd.DataFrame({"Gender Total": gender_count, 
                              "Gender Percentage": gender_percentage})
gender_summary
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Define Table values to summarize Purchase Data by user
purchase_count = purchase_data["Gender"].value_counts()
gender_group = purchase_data.groupby(["Gender"])
price_sum = gender_group["Price"].sum()
avg_price = gender_group["Price"].mean()
avg_total_purchases = (price_sum/purchase_count)

Player_Stats = pd.DataFrame({"Purchases Count": purchase_count,
                            "Average Purchase Price": avg_price,
                            "Total Purchase Value": price_sum,
                            "Avg Total Purchase per Person": avg_total_purchases})

Player_Stats
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
#Find the range of values to base bins
print(purchase_data["Age"].max())
print(purchase_data["Age"].min())
```

```python
#Create bins and labels
bins = [0,9,14,19,24,29,34,39,44]
group_labels = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

#Slice data
pd.cut(purchase_data["Age"],bins, labels=group_labels).head()
```

```python
#add age group to purchase_data
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels=group_labels)
purchase_data.head()
```

```python
#pull data for use
age_df = purchase_data[["SN","Age","Age Group"]]

#drop duplicates
age_df.drop_duplicates(subset="SN",keep='first', inplace=False)

#build elements of age_summary table
age_count = age_df["Age Group"].value_counts()
age_percent = ((age_count/player_count)*100)
age_summary = pd.DataFrame({"Total Count": age_count,
                          "Percentage of Players": age_percent})
age_summary
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
age_data = purchase_data.groupby(["Age Group"])
age_total_purch = age_data["Age Group"].count()
avg_purch_price = age_data["Price"].mean()
total_purch = age_data["Price"].sum()
avg_purch_pp = (total_purch/age_total_purch)

age_data_summary = pd.DataFrame({"Purchase Count": age_total_purch, 
                                 "Avg Purchase Price":avg_purch_price, 
                                 "Total Purchases":total_purch, 
                                 "Avg Purchases Per Player":avg_purch_pp})
                              
                              
                                
age_data_summary
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Group purchase data by player SN
spender_group = purchase_data.groupby(["SN"])

#Calculate number of purchases by player
spender_count = spender_group["SN"].count()

#Calculate average price of item pruchased by each player
spender_avg = spender_group["Price"].mean()

#Calculate total of player purchases
spender_total = spender_group["Price"].sum()

spender_summary = pd.DataFrame({"Purchase Count": spender_count,
                               "Average Purchase Price": spender_avg,
                               "Total Purchase Value": spender_total})

spender_summary = spender_summary.sort_values("Purchase Count", ascending=False)

spender_summary.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Pull Item ID, Name, and Price
item_df = purchase_data[["Item ID","Item Name","Price"]]

#Group data by Item ID
item_group = item_df.groupby(['Item ID'])


#Create calculations for Purchase Count
item_count = pd.DataFrame(item_group["Item Name"].value_counts())
item_count = item_count.rename(columns={"Item Name": "Purchase Count"})
item_count = item_count.reset_index()

#Create calculations for Price
price_list = pd.DataFrame(item_group["Price"].unique())

#Create calculations for Total Purchases
purchases_list = pd.DataFrame(item_group["Price"].sum())

#Merge existing DataFrames to one table
item_summary = item_count.merge(price_list, on="Item ID")
item_summary = item_summary.merge(purchases_list, on="Item ID")

#Format column headers
item_summary = item_summary.rename(columns={"Price_x":"Price", "Price_y":"Total Purchase Value"})

#display table
item_summary
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
summary_sort = item_summary.sort_values("Total Purchase Value", ascending=False)
summary_sort.head()
```
